```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight')

from sklearn.ensemble import ExtraTreesRegressor, RandomForestRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.model_selection  import cross_val_score
from sklearn.metrics import mean_squared_error, make_scorer
```

```python
train = pd.read_csv('train_new.csv')
test = pd.read_csv('test_new.csv')
train_ori = pd.read_csv('/run/media/devesh/Hard Disk/Projects/Machinehack-Melanoma Tumor Size Predicition/input/Train.csv')
```

```python
target = train.pop('tumor_size')
```

```python
def RMSE(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

rmse = make_scorer(RMSE, greater_is_better=False)
```

```python
etc = ExtraTreesRegressor(n_jobs = -1, random_state=2)

scores = cross_val_score(etc, train, target, cv = 5, scoring = rmse)
```
